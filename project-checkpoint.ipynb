# NETFLIX STOCK ANALYSIS PROJECT

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df = pd.read_csv("Netflix.csv")
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize' :(10,5)})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

```python
sns.lineplot(x=df.index, y = df['Volume'], label='Volume')
plt.title('Volume of stock v/s Time')
plt.show()
```

```python
df.plot(y=['High', 'Close', 'Open'], title="Netflix Stock Price")
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize = (15,10))
df.groupby(df.index.day).mean().plot(y = 'Volume', ax=ax1, xlabel = 'Day', color = 'green')
df.groupby(df.index.month).mean().plot(y = 'Volume', ax=ax2, xlabel = 'Month', color = 'magenta')
df.groupby(df.index.year).mean().plot(y = 'Volume', ax=ax3, xlabel = 'Year', color = 'black')
```
